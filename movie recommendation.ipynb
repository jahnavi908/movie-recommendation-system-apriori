#### "The dataset was checked for missing and duplicate values. No missing or duplicate records were found."

```python
data = pd.merge(ratings, movies, on="movieId")
data
```

```python
data = data[data["rating"] >= 4]
data
```

## **🔄 Transaction Encoding**

```python
transactions = data.groupby("userId")["title"].apply(list)
transactions
```

```python
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_arr=te.fit_transform(transactions)
c=pd.DataFrame(te_arr)
c
```

```python
basket = pd.DataFrame(te_arr, columns=te.columns_)
basket.head()
```

## **🛒 Frequent Itemset Mining using Apriori**

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets=apriori(basket,min_support=0.05,use_colnames=True)
frequent_itemsets.sort_values("support",ascending=False).head()
```

```python
from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_itemsets,metric="confidence",min_threshold=0.5)
rules.head(5)
```

```python
rules.sort_values("lift", ascending=False).head(10)
```

## **🎥 Movie Recommendation Function**

```python
def recommend(movie_name):

    recs = rules[rules["antecedents"].apply(lambda x: movie_name in x)]

    recs = recs.sort_values("lift", ascending=False)

    

    recommendations = []

    for movies in recs["consequents"]:
        for movie in movies:
            if movie not in recommendations:
                recommendations.append(movie)

    return recommendations[:5]
```

## **✅ Sample Output**

```python
movie=input("enter movie name to recommend").title()
year=int(input("enter release year"))
name=f"{movie} ({year})"
recommend(name)
```
